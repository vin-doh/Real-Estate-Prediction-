```python
import pandas as pd

# Create two DataFrames
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df2 = pd.DataFrame({'A': [1, 2, 3], 'C': ['x', 'y', 'z']})

# Join the DataFrames based on the 'A' column
joined_df = df1.join(df2.set_index('A'), on='A')

print(joined_df)
```

```python
# Create a DataFrame

vindat = {'A': [1,2,3,4,5], 'B': [6,7,8,9,10]}
index = ('1st', '2nd', '3rd', '4th', '5th')

vt = pd.DataFrame(vindat, index)
```

```python
# Define a function to calculate the sum of two values

def sum_values(x, y):
    return x + y
```

```python
# Apply the function to each row of the DataFrame using apply() and axis = 1

vt['Sum'] = vt.apply(lambda row: sum_values(row['A'], row['B']), axis= 1)
print(vt)
```

```python

# Create a DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'Dave'],
        'Age': [25, 30, 35, 40],
        'Height': [160, 170, 165, 180]}
df = pd.DataFrame(data)

# Apply multiple aggregations using agg()
aggregated_data = df.agg({'Age': ['min', 'max'], 'Height': ['mean', 'sum']})

print(aggregated_data)
```
